```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded!")
```

```python
df = pd.read_excel('/content/P2P_Review.xlsx', sheet_name='SW (2)')

# Keep only the useful columns
df = df[['Company', 'Pilar', 'Pilar 2', 'Weight', 'Score', 'Disclosure', 'Peer Rank'].copy()

# Rename for clarity
df.columns = ['Company', 'Pillar', 'Metric', 'Weight', 'Score', 'Disclosure', 'Peer_Rank']

df.head(20)
```

```python
# Drop rows where Company is missing
df = df.dropna(subset=['Company'])

# Check for missing scores
print(df.isnull().sum())

# Fill any missing scores with 0
df['Score'] = df['Score'].fillna(0)

# Check what companies you have
print(df['Company'].unique())
```

```python
# Filter only pillar-level rows (E, S, G) — not sub-metrics
pillar_df = df[df['Pillar'] == 'ESG']

# Bar chart: ESG pillar scores by company
pillar_df.pivot(index='Company', columns='Metric', values='Score').plot(
    kind='bar', figsize=(10, 5), colormap='Greens', title='ESG Pillar Scores by Company'
)
plt.ylabel('Score')
plt.tight_layout()
plt.show()
```

```python
df.to_csv('/content/esg_benchmark_clean.csv', index=False)
print("Saved! Ready to use in Streamlit.")
```
